```python
class Book:
    def __init__(self,title, author, isbn, publication_year,available_copies):
        self.author = author
        self.isbn = isbn
        self.publication_year = publication_year
        self.available_copies = available_copies
        self.title = title

    def check_out(self):
        if self.available_copies > 0:
            self.available_copies -= 1
            print(f"Thanks for Choosing {self.title}, Happy Reading!")
        else:
            print(f"Sorry Currently There is {self.available_copies} copies available.")

    def return_book(self):
        self.available_copies += 1
        print(f"Thanks For Selecting {self.title} Vsist Again!")

    def display_book_info(self):
        print(self.title)
        print(self.author)
        print(self.isbn)
        print(self.isbn)
        print(self.publication_year)
        print(self.available_copies)

Book1 = Book("ikigai", " Hector Garcia", 1234567362728, 2009,3)
# Book1.return_book() # returning a book
Book1.check_out()  # renting a book

# Book1.display_book_info()  # Display book information
```

```python
class Ticket:
    def __init__(self,ticket_id,event_name,event_date,venue,seat_number,price):
        self.ticket_id = ticket_id
        self.event_name = event_name
        self.event_date = event_date
        self.venue = venue
        self.seat_number = seat_number
        self.price = price
        self.is_reserved = False

    def reserve_ticket(self):
        if not self.is_reserved:
            self.is_reserved = True
            print(f"ticket no. {self.ticket_id} and seat no: {self.seat_number} for {self.event_date} is reserved for event {self.event_name} successfuly")
        else:
            print("ticket booked already")

    def cancel_reservation(self):
        if self.is_reserved:
            self.is_reserved = False
            print(f"Your ticket for {self.event_name} ticket no: {self.ticket_id} is canceled successfuly")

        else:
            print("Book Your ticket, we see no reservetion")

    def display_ticket_info(self):
        print("Ticket Information:")
        print("Venue Ticket No: ",self.ticket_id)
        print("Venue Name: ",self.event_name)
        print("Venue Date: ",self.event_date)
        print("Venue: ",self.venue)
        print("Seat No: ",self.seat_number)
        print("â‚¹",self.price)

new_ticket = Ticket(121212317, "Sundance", "20 Jan", "goa", 12, 2311)

# new_ticket.display_ticket_info()
new_ticket.reserve_ticket()
new_ticket.cancel_reservation()
```

```python
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add_item(self, item):
        self.items.append(item)
        print(f'Congratulations your item: {item} added to the cart')

    def remove_item(self, item):
        if len(self.items) == 0:
             print("Your cart is empty already")
        elif item not in self.items:
            print(f'Error: The item "{item}" is not in your cart.')
        else:
            self.items.remove(item)
            print(f'You have successfully removed "{item}" from your cart.')

    def view_cart(self):
        print("Your Cart is: ",self.items)

    def clear_cart(self):
        if len(self.items) == 0:
             print("Your cart is empty already")
        else:
            self.items.remove()



my_cart = ShoppingCart()


my_cart.add_item("Tv")
my_cart.add_item("Laptop")
my_cart.add_item("Phone")
my_cart.view_cart()
my_cart.remove_item("Tv")
```

```python
class Student:
    def __init__(self, name, age, grade, student_id):
        self. name = name
        self.age = age
        self.grade = grade
        self.student_id =student_id
        self. attendance = {}

    def update_attendance(self, date, status):
       self.attendance[date] = status
       print(f"Attendance for {self.name} on {date}: {status}")

    def get_attendance(self):
        return self.attendance

    def get_average_attendance(self):
        if not self.attendance:
            return 0

        total_days = len(self.attendance)
        present_days = sum(status == 'present' for status in self.attendance.values())
        average_percentage = (present_days / total_days) * 100
        return round(average_percentage, 2)

student1 = Student("Ashwin", 27, "X", 21)
student2 = Student("Sam", 27, "X", 21)
student3 = Student("Chris", 27, "X", 21)




results = student3.get_attendance()
student3.update_attendance("2024-01-23", "present")
results2 = student3.get_average_attendance()


print(results2)
```
