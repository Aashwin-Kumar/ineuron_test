```python
class Vehical:
    def __init__(self,make,model,year):
        self.make = make
        self.model = model
        self.year = year

    def display_info(self):
        print(f"Your car brand is {self.make} model no: {self.model} manufactures in {self.year}")

class car(Vehical):
    def __init__(self, make, model, year,fuel_type):
        super().__init__(make, model, year)     # parent class constructor
        self.fuel_type = fuel_type

    def display_info(self):
        super().display_info()
        print(f"Your car brand is {self.make} model no: {self.model} manufactures in {self.year} running on {self.fuel_type}")

car1 = Vehical("Honda", "HD1321", 2021)
car1.display_info()
car2 = car("Toyota", "TY343",2021, "Hybrid")
car2.display_info()
```

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def display_info(self):
        print(f"{self.name} {self.salary}")


class Manager(Employee):
    def __init__(self, name, salary, department):
        super().__init__(name, salary)
        self.department = department

    def display_info(self):
        print(f"{self.name} from {self.department} has salaey of {self.salary}")


class Developer(Employee):
    def __init__(self, name, salary, programming_language):
        super().__init__(name, salary)
        self.programming_language = programming_language

    def display_info(self):
        print(f"{self.name} from {self.programming_language} has salaey of {self.salary}")


Emp1 = Employee("Jack", 24378)
Emp1.display_info()

emp2 = Manager("Ram", 32123, "HR")
emp2.display_info()

emp3 = Developer("Ash", 10, "Python")
emp3.display_info()
```

```python
class Shape:
    def __init__(self,color, border_width):
        self.color = color
        self.border_width = border_width

    def Display_info(self):
        print(f"{self.border_width} and {self.color}")

class Rectangle(Shape):
    def __init__(self, color, border_width, length, width):
        super().__init__(color, border_width)
        self.length = length
        self.width = width

    def Display_info(self):
        return super().Display_info()

class Circle(Shape):
    def __init__(self, color, border_width, radius):
        super().__init__(color, border_width)
        self.radius = radius

    def Display_info(self):
        super().Display_info()
        print("Radius is",self.radius)


shape1 = Shape("blue", 20)
shape1.Display_info()

shape2 = Rectangle("Black", 10,20, 30)
shape2.Display_info()

shape3 = Circle("Pink", 10,33)
shape3.Display_info()
```

```python
class Device:
    def __init__(self,brand, model):
        self.brand = brand
        self.model = model

    def display_info(self):
        print(F'{self.brand} and {self.model}')


class Phone(Device):
    def __init__(self,brand, model,screen_size):
        super().__init__(self, brand, model)
        self.screen_size = screen_size

    def display_info(self):
        return super().display_info()

class Tablet(Device):
    def __init__(self, brand, model, battery_capacity):
        super().__init__(brand, model)
        self.battery_capacity = battery_capacity

    def display_info(self):
        super().display_info()
        print(f"battry capacity is: {self.battery_capacity}")


# model1 = Device("Apple", "Iphone 15")
# model1.display_info()


model2 = Tablet("Google", "Pixel 9", 6000)
model2.display_info()
```

```python
class BankAccount:
    def __init__(self,account_number,balance):
        self.account_number = account_number
        self.balance = balance

    def display_info(self):
        print("Account No:", self.account_number, "Balance is:",self.balance)


class SavingsAccount(BankAccount):
    def __init__(self, account_number, balance,calculate_interest):
        super().__init__(account_number, balance)
        self.calculate_interest = calculate_interest

    def display_info(self):
        return super().display_info()


class CheckingAccount(BankAccount):
    def __init__(self, account_number, balance,deduct_fee):
        super().__init__(account_number, balance)
        self.deduct_fee = deduct_fee

    def display_info(self):
        # super().display_info()
        print("Account No:", self.account_number, "Balance is:",self.balance, "your deduct Fee is:", self.deduct_fee)




user1 = BankAccount(2321111332, 999)
user1.display_info()

user2 = CheckingAccount(1111111, 999999, 787)
user2.display_info()
```
